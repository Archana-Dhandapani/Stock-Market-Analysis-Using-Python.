```python
#Add a new column ‘Trend’ whose values are based on the 'Daily Return'.
```

```python
df = pd.DataFrame(AMZN,columns = ['Daily Return'])
min_value = df.min()

min_value
```

```python
df = pd.DataFrame(AMZN,columns = ['Daily Return'])
max_value = df.max()

max_value
```

```python
def Trend(x):
    if x > -0.150 and x <0:
        return 'Negative'
    elif x==0.00:
        return 'Neutral'
    elif x > 0 and x <= 3:
        return 'Postive'
```

```python
#AMAZON
AMZN['Trend']= AMZN['Daily Return'].count()
AMZN['Trend']= AMZN['Daily Return'].apply(lambda x:Trend(x))
AMZN     
```

```python
#APPLE
AAPL['Trend']= AAPL['Daily Return'].count()
AAPL['Trend']= AAPL['Daily Return'].apply(lambda x:Trend(x))
AAPL
```

```python
#GOOGLE
GOOG['Trend']= GOOG['Daily Return'].count()
GOOG['Trend']= GOOG['Daily Return'].apply(lambda x:Trend(x))
GOOG.head(50)
```

```python
#MICROSOFT
MSFT['Trend']= MSFT['Daily Return'].count()
MSFT['Trend']= MSFT['Daily Return'].apply(lambda x:Trend(x))
MSFT
```

```python
#Visualize trend frequency through a Pie Chart.
```

```python
#AMAZON
AMZN_pie= AMZN.groupby('Trend')
pie_label = ['Positive','Negative','Neutral']
plt.pie(AMZN_pie['Trend'].count(), labels = pie_label, autopct = '%1.1f%%', radius = 1.5)
plt.show()     
```

```python
#APPLE
AAPL_pie= AAPL.groupby('Trend')
pie_label = ['Positive','Negative','Neutral']
plt.pie(AAPL_pie['Trend'].count(), labels = pie_label, autopct = '%1.1f%%', radius = 1.5)
plt.show()
```

```python
#GOOGLE
GOOG_pie= GOOG.groupby('Trend')
pie_label = ['Positive','Negative']
plt.pie(GOOG_pie['Trend'].count(), labels = pie_label, autopct = '%1.1f%%', radius = 1.5)
plt.show()     
```

```python
#MICROSOFT
MSFT_pie= MSFT.groupby('Trend')
pie_label = ['Positive','Negative']
plt.pie(MSFT_pie['Trend'].count(), labels = pie_label, autopct = '%1.1f%%', radius = 1.5)
plt.show()
```

```python
#What was the correlation between the daily returns of different stocks?
```

```python
data=yf.download(Ticker,start,end)
data
```

```python
data['Close']
```

```python
data['Close'].pct_change()
```

```python
Return=data['Close'].pct_change()
```

```python
plt.figure(figsize=(9, 5))
sns.heatmap(Corr, annot=True, cmap="RdYlGn")
plt.figure()
```
